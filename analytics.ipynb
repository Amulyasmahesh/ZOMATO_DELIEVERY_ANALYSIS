```python
df['rate'] = df['rate'].str.split('/').str[0]
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].str.replace(',', '')
df['approx_cost(for two people)'] = pd.to_numeric(df['approx_cost(for two people)'], errors='coerce')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['rate'], bins=10)
plt.title("Rating Distribution")
plt.xlabel("Ratings")
plt.ylabel("Count")

plt.savefig("rating_distribution.png")  # 👈 SAVE HERE
plt.show()
```

```python
df.groupby('online_order')['rate'].mean()
```

```python
df.groupby('online_order')['rate'].mean().plot(kind='bar')
plt.title("Online Order vs Ratings")
plt.xlabel("Online Order")
plt.ylabel("Average Rating")
plt.savefig("online_vs_rating.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.scatterplot(x=df['approx_cost(for two people)'], y=df['rate'])
plt.title("Cost vs Rating")
plt.savefig("cost_vs_rating.png", dpi=300, bbox_inches='tight')
plt.show()
```
